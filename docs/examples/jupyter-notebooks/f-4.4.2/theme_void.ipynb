# `themeVoid()`


`themeVoid()` is a function to create a plot with a completely blank (or "void") background. 
It removes all elements from the plot, leaving the geometry only.

```kotlin
%useLatestDescriptors
%use lets-plot
%use dataframe
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```kotlin
val p = letsPlot(mpg.toMap()) + geomPie() {  fill = "class"; weight = "displ" }
p
```

Apply `themeVoid()`:

```kotlin
p + themeVoid()
```

Faceted plot:

```kotlin
p + facetGrid(x = "year") + themeVoid()
```
